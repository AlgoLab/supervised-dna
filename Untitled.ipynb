Load model and get new model that [Outputs an intermediate layer](https://keras.io/getting_started/faq/#how-can-i-obtain-the-output-of-an-intermediate-layer-feature-extraction)

Then, evaluate the new model in a subset of labeled data and perform the clustering metrics

```python
import tensorflow as tf
from supervised_dna import ModelLoader
```

```python
# 1. Load model and get output of the last layer
loader = ModelLoader()
model  = loader("resnet50_8mers", 
                weights_path="checkpoints/model-02-0.969.hdf5",
                n_outputs=8) # get compiled model from ./supervised_dna/models
```

```python
model.summary()
```

```python
layer_name = 'global_average_pooling2d_1'
intermediate_layer_model = tf.keras.Model(inputs=model.input,
                                       outputs=model.get_layer(layer_name).output)
```

```python
# 2. Load data and evaluate the model to get features
import json 
import numpy as np
from pathlib import Path
from Bio import SeqIO
from supervised_dna.fcgr import FCGR

from supervised_dna.utils import (
    preprocess_seq
)

fcgr = FCGR(8)

preprocessing = lambda npy: np.expand_dims(npy/10. , axis=-1)

path = Path("/home/disco/Github/supervised-dna/data/hCoV-19")
dataset = list(path.rglob("*.fasta"))

# with open("datasets.json", "r") as fp: 
#     datasets = json.load(fp)    
```

```python
for path_fasta in dataset[:1]: 
    fasta = next(SeqIO.parse(path_fasta, "fasta"))
    array_freq = fcgr(sequence=preprocess_seq(str(fasta.seq)))
    # preproceesing (divide by 10) and add channel axis
    input_model = np.expand_dims(array_freq/10. , axis=-1)
    input_model = np.expand_dims(input_model,axis=0)
    
    # make prediction and get predicted clade
    emb = intermediate_layer_model.predict(input_model)
    
```
